```python
import os
import json
from gensim.models.word2vec import Word2Vec
from pathlib import Path
from preprocessing import preprocess_story_for_embeddings
```

```python
# Define project and data paths
PROJECT_ROOT = Path("/Users/tildeidunsloth/Desktop/Thesis")
DATA_DIR = PROJECT_ROOT / "data/cleaned"
EMBEDDINGS_DIR = PROJECT_ROOT / "embeddings"
```

```python
romance_data_path = DATA_DIR / "romance_stories_cleaned.jsonl"
```

```python
# Read JSONL file
with open(romance_data_path, 'r') as f:
    romance_stories = [json.loads(line) for line in f]
```

```python
# Preprocess stories for Word2Vec
sentences_romance = [
    sentence
    for story in romance_stories
    for sentence in preprocess_story_for_embeddings(story['story'])
]
```

```python
sentences_romance
```

```python
# Word2Vec parameters
num_features = 300  # Word vector dimensionality
min_word_count = 5  # Minimum word count
context = 10  # Context window size
downsampling = 10e-5  # Downsample setting for frequent words
```

```python
def train_and_save_models(sentences, output_dir, genre, **word2vec_params):
    model = Word2Vec(sentences=sentences, **word2vec_params)

    model_name = f"{genre}.w2v"
    save_path = output_dir / model_name

    model.save(str(save_path))
```

```python
train_and_save_models(sentences_romance, 
                      EMBEDDINGS_DIR, 
                      "romance", 
                      vector_size=num_features,
                      min_count=min_word_count,
                      window=context,
                      sg=1,  # Use Skip-gram
                      workers=8,
                      negative=10,
                      epochs = 15 )
```
